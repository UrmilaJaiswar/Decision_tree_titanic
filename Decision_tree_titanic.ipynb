```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df["Cabin"]=df["Cabin"].fillna(df['Cabin'].value_counts().index[0])
```

```python
df.info()
```

```python
df["Embarked"]=df["Embarked"].fillna(df['Embarked'].value_counts().index[0])
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_num=df.select_dtypes(["int64","float64"])
df_cat=df.select_dtypes(object)
```

```python
for col in df_cat :
    le=LabelEncoder() #create the object of LabelEncoder class
    df_cat[col]=le.fit_transform(df_cat[col])
```

```python
#concatenation of both dataset df_num and df_cat and hold new dataset df_new
df_new=pd.concat([df_num,df_cat],axis=1)
```

```python
df_new.head()
```

```python
#select X and Y means input and output from dataset df_new
X=df_new.drop("Survived",axis=1)
Y=df_new["Survived"]#target variable
```

```python
#train_test_split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#classification_report inbuilt class
from sklearn.metrics import classification_report
```

```python
#create object of DecisionTreeClassifier class
dt1=DecisionTreeClassifier()  #by default GINI Index
```

```python
#create user defined function
def create_model(model): #passing arguments model : object
    model.fit(X_train,Y_train) #train the model with 70% data
    Y_pred=model.predict(X_test) #test the model
    print(classification_report(Y_test,Y_pred))
    return model
```

```python
#call function
create_model(dt1)
```

```python
#tree draw
from sklearn import tree
```

```python
features=X.columns
print("Input variables : ",features)
```

```python
fig=plt.figure(figsize=(23,20)) 
#23 width and 20 height :define graph's page size
_=tree.plot_tree(dt1,feature_names=features,filled=True)
#tree_plot() inbuilt function of tree class
```

```python
#create object of DecisionTreeClassifier() and passing argument max_depth
#max_depth can not be more than 8 , can be <=8
```

```python
#create object of DecisionTreeClassifier class
dt2=DecisionTreeClassifier(max_depth=3)  #by default GINI Index
```

```python
#call function
create_model(dt2)
```
